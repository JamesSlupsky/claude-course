```python
# LOAD ENV VARIABLES
from dotenv import load_dotenv
load_dotenv()

# CREATE API CLIENT
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"

# MAKE THE REQUEST
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

```
